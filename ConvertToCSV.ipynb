```python
# Map to hold the relationship between symptoms and their prescriptions, including doctor's name and notes
data_for_csv = []
```

```python
# Unique identifier for each combination of prescriptions to ensure uniqueness
unique_prescriptions = defaultdict(list)
```

```python
# Iterate through each patient record
for patient in patients_data:
    patient_id = patient['id']
    notes = patient.get('notes', '')

    # Create a unique key for the set of prescriptions
    prescriptions = tuple(sorted([(rx['medicine_name'], rx['standard_medicine_id']) for rx in patient['Prescription/Rx']]))
    unique_prescriptions[prescriptions].append(patient_id)
    dignosis = tuple(sorted([(rx1['name']) for rx1 in patient['Diagnosis']]))

    # Map symptoms to the unique prescription identifier, including doctor's name and notes
    for symptom in patient['Complains/Symptoms']:
        symptom_name = symptom['name']
        data_for_csv.append({
            'Doctor ID': doctor_id,
            'Symptom': symptom_name,
            'Prescriptions': '; '.join([f"{name} ({type_})" for name, type_ in prescriptions]),
            'Dignosis':dignosis,
            'Notes': notes
        })
```

```python
df = pd.DataFrame(data_for_csv)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Doctor_Ramesh_Devani.csv')
```

```python
df
```

```python
# Encoding categorical variables
le_symptom = LabelEncoder()
df['Symptom_encoded'] = le_symptom.fit_transform(df['Symptom'])

le_medicine = LabelEncoder()
df['Medicine_encoded'] = le_medicine.fit_transform(df['Prescriptions'])

le_notes = LabelEncoder()
df['Notes_encoded'] = le_notes.fit_transform(df['Notes'])

le_dignosis = LabelEncoder()
df['Dignosis_encoded'] = le_dignosis.fit_transform(df['Dignosis'])
```

```python
def data_split(data, ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train, test = data_split(df, 0.4)
```

```python
X_train = train[['Symptom_encoded']].to_numpy()
X_test = test[['Symptom_encoded']].to_numpy()
```

```python
Y_train = train[['Medicine_encoded']].to_numpy()
Y_test = test[['Medicine_encoded']].to_numpy()
```

```python
Z_train = train[['Notes_encoded']].to_numpy()
Z_test = test[['Notes_encoded']].to_numpy()
```

```python
W_train = train[['Dignosis_encoded']].to_numpy()
W_test = test[['Dignosis_encoded']].to_numpy()
```

```python
model_med = RandomForestClassifier(n_estimators=100, random_state=42)
model_med.fit(X_train, Y_train)
```

```python
model_not = RandomForestClassifier(n_estimators=100, random_state=42)
model_not.fit(X_train, Z_train)
```

```python
model_dig = RandomForestClassifier(n_estimators=100, random_state=42)
model_dig.fit(X_train, W_train)
```

```python
new_symptom_encoded = le_symptom.transform(['PAIN AT VERGE'])

predicted_medicine_encoded = model_med.predict([new_symptom_encoded])
predicted_notes_encoded = model_not.predict([new_symptom_encoded])
predicted_dignosis_encoded = model_dig.predict([new_symptom_encoded])

predicted_medicine = le_medicine.inverse_transform(predicted_medicine_encoded)
predicted_notes = le_notes.inverse_transform(predicted_notes_encoded)
predicted_dignosis = le_dignosis.inverse_transform(predicted_dignosis_encoded)
```
